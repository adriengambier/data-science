# Machine Learning with Linear Regression

In this notebook, we will use an insurance dataset available [here](https://github.com/stedy/Machine-Learning-with-R-datasets/blob/master/insurance.csv). In this dataset, we have different information about the client and the goal is to predict the charges they are going to pay.

For this, we will first go through the dataset using EDA. We will then built a linear regression model and see how close we are with our predictions to the targeted values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

# Data

The dataset contains the following features :
* 'age' : Age of the customer in years
* 'sex' : Sex of the customer ('female' / 'male')
* 'bmi' : Body mass index, ratio of weight to height (kg / m ^ 2)
* 'children' : Number of children
* 'smoker' : Whether or not the customer smokes ('yes' / 'no')
* 'region' : Area in the US where the customer lives ('northeast', 'southeast', 'southwest', 'northwest')
* 'charges' : Medical costs billed by health insurance

```python
insurance_data = pd.read_csv('../datasets/insurance.csv')
```

```python
insurance_data.head()
```

```python
insurance_data.info()
```

No missing data in the dataset and a total of 1338 samples. 

# Exploratory Data Analysis

Now, let's dive in the data analysis. The first thing that would seem highly correlated to medical costs is the age. Indeed, the older you are, the more likely you are to get health issues. So we can plot medical costs against age and see if there is a trend.

```python
sns.lmplot(x = 'age', y='charges', data=insurance_data, fit_reg = False, height=7, aspect = 2)
plt.title('Impact of Age on Charges', fontsize=18)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

We can see a positive correlation between the age and the charges. But there also seems to be 3 different curves so there must be another factor to explain it. Smoking could be the reason since it is the cause of many cancers amongst other health issues. Let's add the 'smoker' category to this plot.

```python
sns.lmplot(x = 'age', y='charges', data=insurance_data, fit_reg = False, hue='smoker', height=7, aspect = 2)
plt.title('Impact of Smokers on Charges', fontsize=18)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

We can see the top curve being people that smoke and the bottom one those that don't smoke. However, the middle one is not very clear. Let's explore other features and get a deeper understanding of the different profiles. For example, the location that could explain the difference between people with similar features.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='charges', data=insurance_data, ci = None, palette='Set2')
plt.title('Impact of Regions on Charges', fontsize=18)
plt.xlabel('Region', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

```python
min_charges = insurance_data.groupby('region').mean()['charges'].min()
max_charges = insurance_data.groupby('region').mean()['charges'].max()

print(f'Highest difference between charges : {(max_charges - min_charges):.2f}')
print(f'Difference between the lowest and highest in percentage : {((max_charges - min_charges)/min_charges)*100:.2f}%')
```

The difference between regions can go up to 2388 which means 20 % more charges depending on regions ! Note that charges on the east seems to be a bit higher than on the west. Now, the bmi (body mass index) which can potentially indicate an underweight or overweight. 

```python
sns.jointplot(x = 'bmi', y='charges', data=insurance_data, height = 10)
```

Again, there is a trend between the bmi and the charges. As the bmi increases, the number of people paying high medical bills increases as well. The bmi can be divided into 4 categories : 'underweight', 'normal', 'overweight' and 'obese'. 

```python
insurance_data.loc[insurance_data['bmi'] <= 18.5, 'bmi_type'] = 'underweight'
insurance_data.loc[(insurance_data['bmi'] > 18.5) & (insurance_data['bmi'] <= 25), 'bmi_type'] = 'normal'
insurance_data.loc[(insurance_data['bmi'] > 25) & (insurance_data['bmi'] <= 30), 'bmi_type'] = 'overweight'
insurance_data.loc[insurance_data['bmi'] > 30, 'bmi_type'] = 'obese'
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='bmi_type', y='charges', data = insurance_data, ci = None, order = ['underweight', 'normal', 'overweight', 'obese'])
plt.title('Mean charges using BMI types', fontsize=18)
plt.xlabel('Body Mass Index', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

While the first 3 categories have similar charges (and surprisingly being underweight is when you pay the less charges), there is a clear distinction with the last category 'obese' in terms of charges. 

There seems to be two different trends on the BMI scatter plot. Let's add the smokers to see if that explains it

```python
ax = sns.lmplot(x = 'bmi', y='charges', data=insurance_data, fit_reg = False, hue='smoker', height=7, aspect = 2)
plt.title('Charges with BMI and smokers', fontsize=18)
plt.xlabel('Body Mass Index', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

The majority non-smokers have similar charges while there is an upward trend for smokers. So it means that BMI has real impact only on people that smoke, and that impact only gets bigger with the BMI increasing.

```python
plt.figure(figsize=(10, 6))

sns.distplot(insurance_data[insurance_data['sex'] == 'male']['charges'],  kde=True, label = 'Men', hist = False)
sns.distplot(insurance_data[insurance_data['sex'] == 'female']['charges'],  kde=True, label = 'Women', hist = False)

plt.title('Charges distribution between genres', fontsize=18)
plt.xlabel('Charges', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.legend(prop={'size': 15})
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x = 'bmi_type', y = 'charges', data = insurance_data, hue = 'sex', order = ['underweight', 'normal', 'overweight', 'obese'], palette = 'Set1')
plt.title('Charges by BMI and sex', fontsize=18)
plt.xlabel('BMI type', fontsize=15)
plt.ylabel('Charges', fontsize=15)
plt.legend(prop={'size': 15})
```

With the charges distribution, we can see that women tend to pay less charges than men. With the boxplot above, we can get a potential reason for that. Underweight is more common for women while overweight and obese is in majority for men. And as we saw earlier, a higher bmi means higher charges

Finally, let's look at the number of children and see the correlation with charges. We first look at the number of children per person.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='children', data=insurance_data, palette='Blues_d')
plt.title('Distribution of the number of children', fontsize=18)
plt.xlabel('Number of Children', fontsize=15)
plt.ylabel('Count', fontsize=15)
```

As expected, few people have lots of children while the majority has 0 or 1 child. The number of people who have 4 or 5 children is very low and it is not representative of the category. To compensate for that, we can create a new category '3+' that will combine people with 3, 4 or 5 children.

```python
insurance_data.loc[insurance_data['children'] == 0, 'children_cat'] = '0'
insurance_data.loc[insurance_data['children'] == 1, 'children_cat'] = '1'
insurance_data.loc[insurance_data['children'] == 2, 'children_cat'] = '2'
insurance_data.loc[insurance_data['children'] > 2, 'children_cat'] = '3+'

insurance_data.drop(['children'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='children_cat', data=insurance_data, palette='Blues_d', order = ['0', '1', '2', '3+'])
plt.title('Distribution of the number of children (Balanced)', fontsize=18)
plt.xlabel('Number of Children', fontsize=15)
plt.ylabel('Count', fontsize=15)
```

200 rows for the category '3+' is a bit more representative. Now, we need to know if there is a difference in the charges depending on the number of children

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="children_cat", y="charges", data=insurance_data, ci = None, palette = 'Set2', order = ['0', '1', '2', '3+'])
plt.title('Impact of the number of children on charges', fontsize=18)
plt.xlabel('Number of Children', fontsize=15)
plt.ylabel('Charges', fontsize=15)
```

Having only one child or none at all is when people pay less charges while having 2 children or more causes the medical costs to increase. An interesting factor to take into consideration when predicting the amount of charges.

# Linear Regression

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

### Encoding

```python
insurance_data = pd.concat([insurance_data, pd.get_dummies(insurance_data['region'])], sort=True, axis=1).drop(['region'], axis=1)
```

```python
insurance_data = pd.concat([insurance_data, pd.get_dummies(insurance_data['children_cat'], prefix='children')], sort=True, axis=1).drop(['children_cat'], axis=1)
```

```python
insurance_data = pd.concat([insurance_data, pd.get_dummies(insurance_data['bmi_type'], prefix='bmi')], sort=True, axis=1).drop(['bmi_type'], axis=1)
```

```python
l_encoder = LabelEncoder()
```

```python
insurance_data['sex'] = l_encoder.fit_transform(insurance_data['sex'])

insurance_data['smoker'] = l_encoder.fit_transform(insurance_data['smoker'])
```

```python
insurance_data
```

### Split train-test

```python
X = insurance_data.drop(['charges'], axis=1)
y = insurance_data[['charges']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Normalization

```python
pd.options.mode.chained_assignment = None
```

```python
age_scaler = StandardScaler().fit(X_train[['age']])
X_train['age'] = age_scaler.transform(X_train[['age']])
X_test['age'] = age_scaler.transform(X_test[['age']])

bmi_scaler = StandardScaler().fit(X_train[['bmi']])
X_train['bmi'] = bmi_scaler.transform(X_train[['bmi']])
X_test['bmi'] = bmi_scaler.transform(X_test[['bmi']])

charges_scaler = StandardScaler().fit(y_train[['charges']])
y_train['charges'] = charges_scaler.transform(y_train[['charges']])
y_test['charges'] = charges_scaler.transform(y_test[['charges']])
```

### Training

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

lin_reg.score(X_test, y_test)
```

The first model we use is a linear regression with the features shown above. We then compute the r2 score of our model, a score of 0 corresponds to a constant value being predicted while a value of 1 is the exact value for each prediction. In our case, we get a score of 0.77 which is not bad for a first try. Now let's try to add some complexity to the data by adding polynomial features.

```python
x_poly = PolynomialFeatures(degree = 2).fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(x_poly, y, random_state = 0)

lin_reg_poly = LinearRegression().fit(X_train, y_train)
lin_reg_poly.score(X_test, y_test)
```

```python
plt.figure(figsize=(8, 8))

plt.scatter(x = lin_reg_poly.predict(X_train), y=y_train, color = 'lightblue')
plt.scatter(x = lin_reg_poly.predict(X_test), y=y_test, color = 'royalblue')

# optimal results
x = np.linspace(0, 60000)
plt.plot(x, x, linewidth = 3, color = 'tomato')

plt.legend(['optimal', 'train', 'test'])

plt.xlabel('Predicted values', fontsize=15)
plt.ylabel('Actual values', fontsize=15)
```

By adding new features, we can see an increase in the score of 0.13. In the scatter plot above, we can still see outliers where the predicted values is lower than the actual value by a large margin but for a simple linear regression, results already look promising.
